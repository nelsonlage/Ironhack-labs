## Import libraries

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
````

## Read the data

````python
breast_cancer = pd.read_csv('breast-cancer.csv', header=None)
breast_cancer.head()
````

````python
# Rename the columns
total_cols = ['id', 'outcome', 'time', 'cell_1_radius', 'cell_1_texture', 'cell_1_perimiter',
              'cell_1_area', 'cell_1_smoothness', 'cell_1_compactness', 'cell_1_concavity',
              'cell_1_concave_points', 'cell_1_symmetry', 'cell_1_fractal_dimension',
              'cell_2_radius', 'cell_2_texture', 'cell_2_perimiter', 'cell_2_area',
              'cell_2_smoothness', 'cell_2_compactness', 'cell_2_concavity',
              'cell_2_concave_points', 'cell_2_symmetry', 'cell_2_fractal_dimension', 'cell_3_radius', 
              'cell_3_texture', 'cell_3_perimiter', 'cell_3_area', 'cell_3_smoothness',
              'cell_3_compactness', 'cell_3_concavity', 'cell_3_concave_points', 'cell_3_symmetry',
              'cell_3_fractal_dimension', 'tumor_size', 'lymph_status']

breast_cancer.columns = total_cols
````

````python
breast_cancer.head()
````

````python
# Check the target variable
breast_cancer['outcome'].value_counts(normalize=True, sort=True, ascending=True).plot.bar(rot=0, color=['LightCoral', 'LightSeaGreen'])

r_label = breast_cancer['outcome'].value_counts(normalize=True).round(4)['R']
n_label = breast_cancer['outcome'].value_counts(normalize=True).round(4)['N']

plt.title('Proportion of results')
plt.annotate(r_label, xy=(0, 0.2), xytext=(-0.13, 0.18), color='white', weight='bold')
plt.annotate(n_label, xy=(1, 0.7), xytext=(0.88, 0.705), color='white', weight='bold')
plt.show()
````

## X/y split

````python
numerics = breast_cancer.select_dtypes('number')
````

````python
numerics.head(5)
````

````python
X = numerics.drop(['id','time'], axis=1)
y = breast_cancer['outcome']
````

## Train/test split & standardization

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()

scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
````

## PCA

````python
pca = PCA(svd_solver='full')

pca.fit(X_train)

X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)

cols = ['PCA_' + str(i) for i in list(range(1,len(X.columns) + 1))]

X_train_pca = pd.DataFrame(X_train_pca, columns=cols)
X_test_pca = pd.DataFrame(X_test_pca, columns=cols)
````

````python
X_train_pca.head()
````

````python
X_test_pca.head()
````

````python
variances = pca.explained_variance_ratio_
total_variance = np.cumsum(variances)

scree = pd.DataFrame({
    'num_pca': range(1, len(variances) + 1),
    'variance': variances,
    'cum_variance': total_variance
})

scree.head(10)
````

````python
sns.set_style("whitegrid")

fig, ax = plt.subplots(2, 1, figsize=(10,15))
plt.subplots_adjust(wspace=0.09)

sns.lineplot(x="num_pca", y="variance", data=scree, ax = ax[0], color = 'black')
ax[0].set_xticks(range(1,32))
sns.barplot(x="num_pca", y="cum_variance", data=scree, ax = ax[1], color = 'lightblue')
ax[1].patches[4].set_color('steelblue')
ax[1].patches[7].set_color('silver')
ax[1].set_yticks([0.2, 0.4, 0.6, 0.8, 0.9, 1.0])
ax[1].axhline(0.8, ls='--', color='midnightblue')
ax[1].axhline(0.9, ls='--', color='gray')

plt.show()
````

5 components are enough to explain over 81% of the variance of the data! With 8, we can explain around 90% of the variation.

```python
# Print every PC with the coefficients for each column
for comp in range(1,len(pca.components_)+1):
    
    print('Principal component', comp)
    my_list = [(abs(value), index+1, X.columns[index], value) for index, value in enumerate(pca.components_[comp-1,:])]
    my_list.sort(reverse=True)
    my_list = [(elem[1], elem[2], round(elem[3],2)) for elem in my_list]
    print(my_list, '\n')
```

## Logistic regression

We'll apply the model only on the relevant components

````python
X_train = X_train_pca.iloc[:,:8]
X_train.head()
````

````python
X_test = X_test_pca.iloc[:,:8]
X_test.head()
````

````python
clf = LogisticRegression().fit(X_train, y_train)
y_hat = clf.predict(X_test)

print(classification_report(y_test, y_hat, digits=4))
````
